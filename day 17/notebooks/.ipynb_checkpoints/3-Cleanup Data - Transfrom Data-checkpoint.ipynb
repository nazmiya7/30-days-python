```python
import os
import pandas as pd
```

```python
cwd=os.getcwd() # dirname(abspath(__file__))
BASE_DIR=os.path.dirname(cwd)
DATA_DIR=os.path.join(BASE_DIR,'data')
CACHE_DIR=os.path.join(BASE_DIR,'cache')
working_file=os.path.join(CACHE_DIR,'movies-box-office-dataset.csv')
```

```python
df=pd.read_csv(working_file)
```

```python
df.head(n=1200)
```

```python
df['Rank']=-1
```

```python
df.head()
```

```python
df['Dosemtic %']=df['%']
df['Foreign %']=df['%.1']
df.head()
```

```python
df.drop(['%','%.1'],axis=1)
```

```python
df.head()
```

```python
df['Worldwide']=df['Worldwide'].replace("$", "").replace(",", "")
df.head()
```

```python
to_clean_cols=['Worldwide','Domestic','Foreign']

def corrency_str_to_int(current_val):
    corrency_val=current_val.replace("$", "").replace(",", "")
    try:
        currency_val=int(currency_val)
    except:
        # "-" into 0
        currency_val=0
    return currency_val
  

def clean_col(row):
    
    """
    row=pandas series
    dataframe=collection of pandas series
    """
    # print(row)
    for col in to_clean_cols:
        current_val=row[col]
        row[col]=corrency_str_to_int(current_val)
    return row

df_cleaned=df.apply(clean_col,axis=1)
df_cleaned.head()
```
